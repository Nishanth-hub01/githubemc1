```python
class week1():
    def mean(self,lst):
        print("The mean of the list:",sum(lst)/len(lst))

    def median(self,lst):
        if len(lst)%2!=0:
            mid=(len(lst)//2)
            print("The median of the list:",lst[mid])
        else:
            lst.sort()
            first=lst[(len(lst)//2)-1]
            second=lst[len(lst)//2]
            mid=(first+second)/2
            print("The median of the list:",mid) 
            
lst=[10,20,30,40,50] 
w1=week1()
w1.mean(lst)
w1.median(lst)
```

```python
def week2():
    myfile=open("D:/python file demo/universe_essay.txt","r")
    file=myfile.read()
    myfile.close()

    file1=file.replace("\n\n"," ")
    file2=file1.replace("\n"," ")
    lt=file2.split(" ")
    #print(len(lt))

    lt1=lt.remove("")

    suf=".,:;"
    lta=[]
    for i in lt:
        if i[-1] in suf:
            i=i.removesuffix(i[-1])
        lta.append(i)
    #print(lta) 

    import pandas as pd
    #print(pd.value_counts(lta))
    k=list(pd.Series(lta).value_counts().items())
    for i in k:
        print(i)

week2()    
```

```python
import os
import re
import shutil
def week3():
    source="C:\\Users\\Nishanth\\Downloads"

    dic={"Documents":[".doc",".docx",".pdf",".xlsx",".odt",".rtf",".txt",".md"],"Pictures":[".jpg",".png",".gif",".bmp",".tiff",".webp",".heif"],"Music":[".mp3",".wav",".aac",".flac",".ogg",".m4a",".opus",".wma"],"Videos":[".mp4",".mkv",".avi",".mov",".wmv",".3gp",".m4v"]}
    for file in os.listdir(source):
        destination="C:\\Users\\Nishanth"
        for des_folder,des_exe in dic.items():
            for exe in des_exe:
                if file.endswith(exe):
                    old_file=os.path.join(source,file)
                    new_file=os.path.join(destination,des_folder)
                    print(new_file)
                    shutil.move(old_file,new_file)
                    print(file,"is moved")
                
week3()                
```
